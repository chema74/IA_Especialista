## 3. Normalización y Exportación Final
Realizamos una limpieza de strings para asegurar la uniformidad de los datos:
1. **Strip**: Eliminamos espacios accidentales al inicio o final.
2. **Lower**: Convertimos todo a minúsculas.
3. **Exportación**: Guardamos el tercer y último búnker de la semana.

```python
import os

# 1. Función para encontrar la raíz del proyecto dinámicamente
def buscar_raiz(nombre_raiz="IA_Especialista"):
    ruta = os.getcwd()
    while True:
        if os.path.basename(ruta) == nombre_raiz:
            return ruta
        padre = os.path.dirname(ruta)
        if padre == ruta: # Hemos llegado a la raíz del disco duro (C:\)
            return os.getcwd() # Devolvemos la actual como emergencia
        ruta = padre

# 2. Definimos las rutas basadas en la RAÍZ real
raiz = buscar_raiz()
folder_final = os.path.join(raiz, 'data', 'processed')
file_final = os.path.join(folder_final, 'marketing_limpio.csv')

# 3. Crear la carpeta si no existe (esto asegura que se cree en la raíz)
os.makedirs(folder_final, exist_ok=True)

# 4. Guardar el archivo
df_mkt.to_csv(file_final, index=False)

print(f"📍 Raíz detectada: {raiz}")
if os.path.exists(file_final):
    print(f"✅ ARCHIVO UBICADO CORRECTAMENTE EN: {file_final}")
else:
    print("❌ Error: No se pudo verificar el guardado.")
```
